# Data Analysis for Uber Trip For Sept-14

### Load the relevant libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%pylab inline
```

#### Load raw data

```python
data = pd.read_csv('uber-raw-data-sep14.txt')
```

#### Confirm number of rows and columns

```python
data.shape
```

### Data cleaning

#### Check for the first five rows

```python
data.head()
```

#### Check for missing values

```python
data.isnull().sum()
```

#### Convert datetime to datetime object

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

```python
data
```

#### Creating columns for DoM, Weekday and Hour

```python
def get_dom(dt):
    return dt.day

data['DOM'] = data['Date/Time'].map(get_dom)

def get_weekday(dt):
    return dt.weekday()

data['Weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['Hour'] = data['Date/Time'].map(get_hour)

data.head()
```

### Visualisation

```python
data['DOM'].hist(bins=30, rwidth=.8, range=(0.5, 30.5))
xlabel('Date of the Month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Sept 2014')
```

#### Groupby() function

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('DOM').apply(count_rows)
by_date
```

```python
bar(range(1, 31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1, 31), by_date_sorted)
xticks(range(1, 31), by_date_sorted.index)
xlabel('Date of the Month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Sept 2014');
```

```python
data['Hour'].hist(bins=20, rwidth=.8, range=(0.5, 24))
xlabel('Hour')
ylabel('Frequency')
title('Frequency by Hour - Uber - Sept 2014')
```

```python
hist(data.Weekday, bins=7, rwidth=.8, range=(-.5, 6.5), color='blue',alpha=.5)
xticks(range(7), 'Mon Tue Wed Thur Fri Sat Sun'.split())
xlabel('Weekday')
ylabel('Frequency')
title('Frequency by Mon - Uber - Sept 2014')
```

```python
by_h_w = data.groupby('Hour Weekday'.split()).apply(count_rows)
by_h_w
```

```python
by_h_w.unstack()
```

```python
by_cross = data.groupby('Weekday Hour'.split()).apply(count_rows).unstack()
```

```python
sns.heatmap(by_cross)
```
