```python
# import modules and libraries
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

#### Use glob() to list all files that match a pattern and sort the results

```python
uber_raw_data = sorted(glob('uber-raw-data-*.txt'))
uber_raw_data
```

#### Use a generator expression to read the files, assign(), to create a new column, and concat() the DataFrame

```python
uber_data = pd.concat((pd.read_csv(data).assign(filename = data)
          for data in uber_raw_data), ignore_index = True)

uber_data.to_csv('all_data.csv', index = False)
```

#### Read in updated DataFrame

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### Data cleaning
#### Confirm number of rows and columns

```python
all_data.shape
```

#### Check for missing data

```python
all_data.isnull().sum()
```

#### Drop filename column

```python
all_data = all_data.drop('filename', axis = 1)
all_data
```

#### Create month column and convert to int

```python
all_data['Month'] = all_data['Date/Time'].str[0]
all_data['Month'] = pd.to_numeric(all_data['Month'])
all_data.head()
```

#### Convert datetime to date time object

```python
all_data['Date/Time'] = pd.to_datetime(all_data['Date/Time'])
all_data
```

#### Creating columns for DoM, Weekday and Hour

```python
def get_dom(dt):
    return dt.day

all_data['DOM'] = all_data['Date/Time'].map(get_dom)

def get_weekday(dt):
    return dt.weekday()

all_data['Weekday'] = all_data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

all_data['Hour'] = all_data['Date/Time'].map(get_hour)

all_data.head()
```

### Exolorationary Data Analysis

#### Summary Statistics

```python
all_data.describe()
```

#### What month had the highest hour of trips? How many hours?

```python
results = all_data.groupby('Month').sum()
results
```

```python
Month = range(1,7)
plt.bar(Month, results['Hour'])
plt.xticks(range(1,7), 'Apr May Jun Jul Aug Sept'.split())
plt.xlabel('Month')
plt.ylabel('Hour')
title('Hour by Months - Uber Tirps 2014')
plt.show()
```

#### What day of the week has the highest hours of trips

```python
results = all_data.groupby('Weekday').sum()
results
```

```python
bases = all_data['Weekday'].unique()
plt.bar(bases, results['Hour'])
plt.xticks(range(7), 'Sun Mon Tue Wed Thu Fri Sat'.split())
plt.xlabel('Weekday')
plt.ylabel('Hour')
title('Hour by Weekday - Uber Tirps 2014')
plt.show()
```

#### What date of the month has the highest hours of trips

```python
results = all_data.groupby('DOM').sum()
results
```

```python
bases = all_data['DOM'].unique()
plt.bar(bases, results['Hour'])
plt.xticks(range(1,32), rotation=90)
plt.xlabel('Date of Month')
plt.ylabel('Hour')
plt.title('Hour by DoM - Uber Tirps 2014')
plt.show()
```

#### What Base had the hihgest hours of Trips

```python
results = all_data.groupby('Base').sum()
results
```

```python
bases = all_data['Base'].unique()
plt.bar(bases, results['Hour'])
plt.xticks(base)
plt.xlabel('Base')
plt.ylabel('Hour')
title('Hour by Bases - Uber Tirps 2014')
plt.show()
```

#### What time is the peak period for trips

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xlabel('Hour')
plt.ylabel('Frequency')
plt.xticks(hours)
plt.grid()
plt.title('Peak Period For Trips')
plt.show()
```

#### What day of the week has the highest number of trips

```python
days = [day for day, df in all_data.groupby('Weekday')]
plt.plot(days, all_data.groupby(['Weekday']).count())
plt.xlabel('Weekday')
plt.ylabel('Frequency')
plt.xticks(range(7), 'Sun Mon Tue Wed Thu Fri Sat'.split())
plt.grid()
plt.title('Frequency by Weekday')
plt.show()
```
