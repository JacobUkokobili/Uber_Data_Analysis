```python
%pylab inline
import pandas as pd
import seaborn as sns
```

# Load CSV file into memory

```python
data = pd.read_csv('uber-raw-data-apr14.txt')
```

```python
data.head()
```

#### Convert datatime and add some useful columns

```python
 data['Date/Time'] = data['Date/Time'].map(pd.to_datetime )
```

```python
data.head()
```

```python
def get_dom(dt):
    return dt.day

data['DOM'] = data['Date/Time'].map(get_dom)

def get_weekday(dt):
    return dt.weekday()

data['Weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['Hour'] = data['Date/Time'].map(get_hour)

data.head()
```

#### Analysis for DoM

```python
# analysing the DoM
data['DOM'].hist(bins=30, rwidth=.8, range=(0.5, 30.5))
xlabel('Date of the Month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Apr 2014')
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('DOM').apply(count_rows)
by_date
```

```python
bar(range(1, 31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1, 31), by_date_sorted)
xticks(range(1, 31), by_date_sorted.index)
xlabel('Date of the Month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Apr 2014')
;
```

#### Analysis for Hour

```python
# analysing the Hour
data['Hour'].hist(bins=20, rwidth=.8, range=(0.5, 24))
xlabel('Hour')
ylabel('Frequency')
title('Frequency by Hour - Uber - Apr 2014')
```

#### Analysis for Hour

```python
# analysing the Hour
hist(data.Weekday, bins=7, rwidth=.8, range=(-.5, 6.5), color='blue',alpha=.5)
xticks(range(7), 'Mon Tue Wed Thur Fri Sat Sun'.split())
xlabel('Weekday')
ylabel('Frequency')
title('Frequency by Mon - Uber - Apr 2014')
```

#### Cross Analysis (Hour, DoW)

```python
by_h_w = data.groupby('Hour Weekday'.split()).apply(count_rows)
by_h_w
```

```python
by_h_w.unstack()
```

```python
by_cross = data.groupby('Weekday Hour'.split()).apply(count_rows).unstack()
```

```python
sns.heatmap(by_cross)
```
